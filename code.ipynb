```python
import requests
import re
import pandas as pd

# Fetch HTML content from the URL
response = requests.get('https://www.montclair.edu/school-of-computing/staff-directory/')

print(response.content)

html_content = response.content.decode('utf-8')

#patterns to fetch full time html code
after_full_time_html = re.findall(r'<h2>Full Time Faculty</h2>(.+)', html_content, re.DOTALL)
full_time_faculty_html = re.findall(r'(.+)<h2>Adjunct Faculty</h2>', after_full_time_html[0], re.DOTALL)

# pattern to fetch name, title, phone, email and location
pattern = r'<p class="name"><a.*?>(.*?)</a></p>.*?<p class="title">(.*?)</p>.*?Phone.*?href="tel:(.*?)">.*?Email.*?href="mailto:(.*?)">.*?Location.*?<dd>(.*?)</dd>'

# fetch full-time faculty details
full_time_faculty_details = re.findall(pattern, full_time_faculty_html[0], re.DOTALL)

# unpack the details into separate lists
name_list, title_list, phone_list, email_list, location_list = zip(*full_time_faculty_details)


# Create DataFrame from extracted information
df = pd.DataFrame({
    'name': name_list,
    'tile': title_list,
    'phone': phone_list,
    'email': email_list,
    'location': location_list
})

display(df)
```
